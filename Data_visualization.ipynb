# Matplotlib

`Matplotlib` is a Python library used for creating charts and graphs.

```python
# Import plotting functions from Matplotlib

from matplotlib import pyplot as plt
```

In `pyplot`, we first create the graph using the `.plot()` function and we then show it using the `show()` function:

```python
x = [val for val in range(-10,11)]
y = [val ** 2 for val in x]
plt.plot(x,y)
plt.show()
```

## Customization

We can add multiple lines to the same plot. To do this, place the several calls to the `plot()` function together before showing the graph with `.show()`

```python
x = [val for val in range(-10,11)]
y = [val ** 2 for val in x]
z = [val ** 3 for val in x]
plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

Additionally, we can customize the lines using different arguments passed to the `.plot()` function. See this [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) to get more information on the topic

```python
x = [val for val in range(-10,11)]
y = [val ** 2 for val in x]
z = [val ** 3 for val in x]
plt.plot(x, y, color = 'blue', linestyle = '--')
plt.plot(x, z, color = 'red', linestyle = ':')
plt.show()
```

To zoom in our plot we can use the `.axis()` function, which takes a list as an argument. This list indicates:

- Minimum x value
- Maximum x value
- Minimum y value
- Maximum y value

```python
x = [x for x in range(101) if x % 2 == 0]
y = [x/2 for x in x]
plt.plot(x,y)
plt.axis([20, 40, 10, 30])
plt.show()
```

Use `.ylabel()`,`.xlabel()` and `.title()` to specify labels for your plot

```python
x = range(12)
y = [3000, 3005, 3010, 2900, 2950, 3050, 3000, 3100, 2980, 2980, 2920, 3010]
plt.plot(x, y)
plt.axis([0, 12, 2900, 3100])
plt.xlabel('Time (years)')
plt.ylabel('Dollars spent on coffee')
plt.title('My Last Twelve Years of Coffee Drinking')
plt.show()
```

## Subplots

Sometimes we want to display two lines side by side, rather than in an overlapping plot. In this case, we would use the `.subplot(nrow, ncol, index)` function. After this, we can use `.plot()` call to incorporate a particular plot to the composition.

```python
# Values
x1 = [val for val in range(-10, 11)]
x2 = [val for val in range(-20, 21)]
y1 = [val ** 2 for val in x1]
y2 = [val ** 2 for val in x2]

# First subplot
plt.subplot(1, 2, 1)
plt.plot(x1, y1)
plt.title('First plot')

# Second subplot
plt.subplot(1, 2, 2)
plt.plot(x2, y2)
plt.title('Second plot')

# Show subplots
plt.show()
```

We can use the `.subplots_adjust()` function to adjust the spacing between our subplots. `kwargs` for this function include:

- `left`
- `right`
- `bottom`
- `top`
- `wspace` (horizontal space between adjacent plots)
- `hspace` (vertical space between adjacent plots)

```python
x = range(7)
straight_line = [0, 1, 2, 3, 4, 5, 6]
parabola = [0, 1, 4, 9, 16, 25, 36]
cubic = [0, 1, 8, 27, 64, 125, 216]

plt.subplot(2, 1, 1)
plt.plot(x, straight_line)

plt.subplot(2, 2, 3)
plt.plot(x, parabola)

plt.subplot(2, 2, 4)
plt.plot(x, cubic)

plt.subplots_adjust(wspace=0.35, bottom=0.2)
plt.show()
```

## Legends

We can label the multiple lines on a graph using the `plt.legend()` function

```python
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64])
plt.legend(['parabola', 'cubic'])
plt.show()
```

This function can also take a keyword argument, `loc` which will position the legend on the figure

We can also specify the label inside the `plt.plot()` call

```python
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16],
         label="parabola")
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64],
         label="cubic")
plt.legend() # Still need this command!
plt.show()
```

## Modifying ticks

In order to modify ticks, we first need to save our subplot to an `axes` object. Then, we can operate over it with the `.set_xticklabels()` function

```python
from matplotlib import pyplot as plt
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

ax1 = plt.subplot(1,1,1)
x_values = [i for i in range(len(months))]
plt.plot(x_values, visits_per_month)
plt.xlabel('Month')
plt.ylabel('Total visits')
ax1.set_xticklabels(months)
plt.show()
```

## Figures

Sometimes, we might want to have some more control over the size of our figure. We can do this by creating a new figure window with the `plt.figure()` function. After calling it, we can use `plt.plot()` to plot the data inside this blank figure. Finally, we can save the figure by using `plt.savefig('name.png')`.

Before plotting a new plot, it is important to close the previous ones. We can do this by using `plt.close('all')`

```python
x = [i for i in range(11)]
y = [i ** 2 for i in x]

# Open a new figure

plt.figure(figsize=(10, 10))
plt.plot(x, y)

# Save it

plt.savefig('myfig.png')
plt.close('all')
```

## Practical example

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]

persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]

blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]


# create your figure here
plt.figure(figsize=(12, 8))

ax1 = plt.subplot(1,2,1)
x_values = [i for i in range(len(months))]
plt.plot(x_values, visits_per_month)
plt.xlabel('Month')
plt.ylabel('Total visits')
plt.title('Monthly visits')
ax1.set_xticklabels(months)

ax2 = plt.subplot(1,2,2)
plt.plot(x_values, key_limes_per_month, color='green')
plt.plot(x_values, persian_limes_per_month, color='pink')
plt.plot(x_values, blood_limes_per_month, color='red')
plt.legend(['Keylimes', 'Persian limes', 'Blood limes'])
ax2.set_xticklabels(months)
plt.xlabel('Month')
plt.ylabel('Number of limes sold')
plt.title('Monthly sales by lime type')
plt.show()
plt.savefig('Sublime-lime.png')
```
